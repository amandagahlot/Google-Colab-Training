Example

```python
#Change to Z scores
df['ZphysQOL'] = zscore(df['physQOL'])
df['ZemoQOL'] = zscore(df['emoQOL'])
df['ZTotalFrSBE'] = zscore(df['TotalFrSBE'])
df['ZEngagement'] = zscore(df['Engagement'])
df['ZSPS'] = zscore(df['SPS'])
df['ZcogQOL'] = zscore(df['cogQOL'])
```

```python
#Build the figure
fig, ax_lst = plt.subplots(2,2)
ax_lst[0,0].set_title("Emotional")
ax_lst[0,1].set_title("Cognitive")
ax_lst[1,0].set_title("Physical")
ax_lst[1,1].set_title("Self-Regulation")

# Panel 1: ZEngagement vs. ZemoQOL
sns.scatterplot(x='ZemoQOL', y='ZEngagement', data=df, ax=ax_lst[0, 0])

# Panel 2: ZEngagement vs. ZcogQOL
sns.scatterplot(x='ZcogQOL', y='ZEngagement', data=df, ax=ax_lst[0, 1])

# Panel 3: ZEngagement vs. ZSPS
sns.scatterplot(x='ZphysQOL', y='ZEngagement', data=df, ax=ax_lst[1, 0])

# Panel 4: ZEngagement vs. ZTotalFrSBE
sns.scatterplot(x='ZTotalFrSBE', y='ZEngagement', data=df, ax=ax_lst[1, 1])

plt.tight_layout() #improve visual
plt.show()
```

Median Split

```python
median = df['TimeSinceInjury'].median()

# split the variable into two groups based on the median
df['group'] = df['TimeSinceInjury'].apply(lambda x: 'early' if x <= median else 'late')
```

Bar plot based on median split

```python
#example
colors = ["#0000FF", "#FF0000"]

fig, ax = plt.subplots(1,1, figsize=(4,6))
sns.barplot(x='group', y="ZEngagement", data=df, palette=colors, ax=ax)
ax.set_xticklabels(['Early Recovery', 'Late Recovery'])
ax.set(ylabel='Re-Engagement', xlabel='Time since injury')
plt.tight_layout()
```

```python
#Off to the races
import numpy as np
import matplotlib.pyplot as plt

y = np.array([32, 14, 15, 6, 6, 6, 6, 6, 3, 3, 3])
labels = ["Satisfaction", "Use Behavior", "Acceptance", "Uptake", "Willingness", "Participation", "Adherence", "Advantages & Disadvantages", "Benefits", "Challenges & Suggestions", "Requirements"]

#color palette
num_of_data = len(y)
colors = plt.cm.tab20(np.linspace(0, 1, num_of_data))

# Using labeldistance to draw labels outside the pie with lines connecting them
plt.pie(y, startangle=180, colors=colors, labels=labels, autopct='%1.1f%%', pctdistance=0.8, labeldistance=1.1)

# Title for the pie chart
plt.title("Constructs of Interest", weight='bold')

plt.tight_layout() #looks better when you download
plt.savefig('ConstructsofInterest.tif', format="tiff", dpi=300, bbox_inches='tight') #this saves it right to your drive and then you can download it to your computer in the correct format
plt.show() #displays image (pro tip: if you put plt.show before plt.savefig, it won't save)
```
